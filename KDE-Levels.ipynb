```python
from analysis_tools import *
import bokeh
from bokeh.plotting import *
output_notebook()
```

```python
runs = runsFromFolder('kde_level_testing')
```

```python
plot = figure(title="Motor Temp")

for run in runs:
    spike = run.data
    tempSpike = [x.motorTemp for x in spike]
    timeSpike = [x.time for x in spike]

    plot.line(timeSpike, tempSpike)
        
show(plot)
```

```python
props = list(filter(lambda x: not '_' in x, dir(runs[0].data[0])))

for prop in props:
    plot = figure(title=prop)

    for run in runs:
        spike = run.data
        tempSpike = [getattr(x, prop) for x in spike]
        timeSpike = [x.time for x in spike]

        plot.line(timeSpike, tempSpike)
        
    show(plot)
```
