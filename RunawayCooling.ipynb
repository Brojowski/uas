```python
from analysis_tools import *

import bokeh
from bokeh.plotting import *
from bokeh.models import LinearAxis, Range1d, Legend
output_notebook()
```

```python
unfilteredRuns = runsFromFolder('data_and_control')
```

```python
# Filter runs here
#runs = runsUnfiltered
runs = [r for r in filter(lambda r: 'Runaway-KDE' in r.file, unfilteredRuns)]
```

```python
for run in filter(lambda x: len(x.data) > 100 , runs):
    time = [x.time for x in run.data]
    temp = [x.motorTemp for x in run.data]
    pwm = [x.pwm for x in run.data]
    
    plot = figure(title=run.file)
    plot.line(x=time, y=temp)
    plot.line(time, pwm)
    
    plot.extra_y_ranges = {"foo": Range1d(start=1000, end=2100)}
    plot.add_layout(LinearAxis(y_range_name="foo", axis_label='Throttle'), 'right')
    
    show(plot)
```

```python
def plotWithThrottle(run, bottomOfPlot, topOfPlot, color):
    plot = figure(title='Time vs Temp and Throttle', y_range=[bottomOfPlot, topOfPlot], x_axis_label='Time (s)', y_axis_label='Motor Temperature (F)')
    

    time = [x.time for x in run.data]
    temp = [x.motorTemp for x in run.data]
    pwm = [(topOfPlot-bottomOfPlot)*((x.pwm - 1000) / (2000-1000) ) + bottomOfPlot for x in run.data]
    
    l1 = plot.line(x=time, y=temp, line_color=color)
    l2 = plot.line(time, pwm, line_color='grey')
    
    plot.extra_y_ranges = {"foo": Range1d(start=0, end=100)}
    plot.add_layout(LinearAxis(y_range_name="foo", axis_label='Throttle (%)'), 'right')
    
    legend = Legend(items=[ ("Motor Temp" , [l1]), ("Throttle" , [l2])])
    plot.add_layout(legend, 'center')
    return plot
```

```python
i = 0
colors = ['red', 'green', 'blue']

plot = figure(title=run.file, y_range=[bottomOfPlot, topOfPlot], x_axis_label='Time (s)', y_axis_label='Motor Temperature (F)')

for run in filter(lambda x: len(x.data) > 100 , runs):

    plot = plotWithThrottle(run, 82, 141, colors[i])

    i += 1


show(plot)
```
